```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load and investigate the data here:
tennis = pd.read_csv("tennis_stats.csv")
print(tennis.head())
```

```python
# perform exploratory analysis here:
plt.scatter(tennis.Aces, tennis.Winnings)
plt.show()
plt.clf()
plt.scatter(tennis.ReturnPointsWon, tennis.Winnings)
plt.show()
plt.clf()
plt.scatter(tennis.BreakPointsConverted, tennis.Winnings)
plt.show()
plt.clf()
plt.scatter(tennis.BreakPointsOpportunities, tennis.Winnings)
plt.show()
plt.clf()
plt.scatter(tennis.Wins, tennis.Winnings)
plt.show()
plt.clf()
```

```python
features_3 = tennis[['Wins']]
outcome_3 = tennis[['Winnings']]
features_train3, features_test3, outcome_train3, outcome_test3 = train_test_split(features_3, outcome_3, train_size = 0.8)
```

```python
model = LinearRegression()
model.fit(features_train3,outcome_train3)
model.score(features_test3,outcome_test3)
prediction = model.predict(features_test3)
plt.scatter(outcome_test3, prediction, alpha=0.4)
plt.show() 
plt.clf()
```

```python
## perform two feature linear regressions here:
features_4 = tennis[['BreakPointsOpportunities', 'Wins']]
outcome_4 = tennis[['Winnings']]
features_train4, features_test4, outcome_train4, outcome_test4 = train_test_split(features_4, outcome_4, train_size = 0.8)
```

```python
model = LinearRegression()
model.fit(features_train4,outcome_train4)
model.score(features_test4,outcome_test4)
prediction = model.predict(features_test4)
plt.scatter(outcome_test4, prediction, alpha=0.4)
plt.show() 
plt.clf()
```
